```
def extract(file, target_dir):
    if (os.path.exists(target_dir) == False):
      print(f'{target_dir} does not exist. Creating now...')
      os.mkdir(target_dir)
    else:
      print("Path exists")

    from zipfile import ZipFile
    with ZipFile(f'{DATA_DIR}{os.sep}{file}', 'r') as zf:
        zf.extractall(target_dir)
    print('Extracted files.')
```

```
extract('train_label_20k.zip', TARGET_DIR)
```
